```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
# Load dataset
dataset = pd.read_csv('dataset/train.csv')
X = dataset.iloc[:, 2:59].values
y = dataset.iloc[:, 1].values
```

```python
columns = dataset.columns
columns.sort_values
```

```python
#dataset.describe().transpose()
```

```python
x = np.random.normal(size=100)
sns.distplot(x);
sns.distplot(x, kde=False, rug=True);
```

```python
# categorical vars
categorical_vars = []
for c in columns:
    if c.endswith('cat'):
        print(c)
        print(dataset[c].unique())
        print(dataset[c].value_counts())        
        categorical_vars.append(c)
        #sns.distplot(dataset[c], kde=False, rug=True);
print(categorical_vars)
categorical_ds = dataset.filter(items=categorical_vars)
```

```python
# binary vars
binary_vars = []
for c in columns:
    if c.endswith('bin'):
        binary_vars.append(c)
print(binary_vars)

binary_ds = dataset.filter(items=binary_vars)
binary_ds
```

```python
import re
# general vars
general_vars = []
for c in columns:
    if(re.match('.*?([0-9]+)$', c)):
        general_vars.append(c)
        
print(general_vars)
general_ds = dataset.filter(items=general_vars)
general_ds
```

```python
# float vars
float_vars =  dataset.select_dtypes(include=['float64'])
float_vars.sort_values
float_vars
```
